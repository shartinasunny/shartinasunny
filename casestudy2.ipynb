# 3. Who scores the most on average for math, reading and writing based on
      ● Gender
      ● Test preparation course

```python
fem_math_mean=dff['math score'].mean()
mal_math_mean=dfm['math score'].mean()
fem_read_mean=dff['reading score'].mean()
mal_read_mean=dfm['reading score'].mean()
fem_write_mean=dff['writing score'].mean()
mal_write_mean=dfm['writing score'].mean()

print('Average Male Scores in:\n\n Maths:',mal_math_mean,'\nReading:',mal_read_mean,'\nWriting:',mal_write_mean)
print('\nAverage Female Scores in:\n\n Maths:',fem_math_mean,'\nReading:',fem_read_mean,'\nWriting:',fem_write_mean)

dfn= data[(data['test preparation course']=='none')]
dfc= data[(data['test preparation course']=='completed')]

non_math_mean=dfn['math score'].mean()
com_math_mean=dfc['math score'].mean()
non_read_mean=dfn['reading score'].mean()
com_read_mean=dfc['reading score'].mean()
non_write_mean=dfn['writing score'].mean()
com_write_mean=dfc['writing score'].mean()

print('\n\n\nAverage for test preparation course not done Scores in:\n\n Maths:',non_math_mean,'\nReading:',non_read_mean,'\nWriting:',non_write_mean)
print('\nAverage for test preparation course completed Scores in:\n\n Maths:',com_math_mean,'\nReading:',com_read_mean,'\nWriting:',com_write_mean)
```

# insights
Male scored more in maths and rest are females and Based on test preparation course, The ones who completed has a higher average in all three sections which is maths, reading and writing which are 69.97, 74.18 and 74.68 respectively.

# 4. What do you think about the scoring variation for math, reading and writing based on

      ● Test preparation course

```python
fem_math_std=dff['math score'].std()
mal_math_std=dfm['math score'].std()
fem_read_std=dff['reading score'].std()
mal_read_std=dfm['reading score'].std()
fem_write_std=dff['writing score'].std()
mal_write_std=dfm['writing score'].std()

print('Deviation of Male Scores in:\n\n Maths:',mal_math_std,'\nReading:',mal_read_std,'\nWriting:',mal_write_std)
print('\nDeviation of Female Scores in:\n\n Maths:',fem_math_std,'\nReading:',fem_read_std,'\nWriting:',fem_write_std)


non_math_std=dfn['math score'].std()
com_math_std=dfc['math score'].std()
non_read_std=dfn['reading score'].std()
com_read_std=dfc['reading score'].std()
non_write_std=dfn['writing score'].std()
com_write_std=dfc['writing score'].std()

print('\n\n\nDeviation for test preparation course not done Scores in:\n\n Maths:',non_math_std,'\nReading:',non_read_std,'\nWriting:',non_write_std)
print('\nDeviation for test preparation course completed Scores in:\n\n Maths:',com_math_std,'\nReading:',com_read_std,'\nWriting:',com_write_std)
```
