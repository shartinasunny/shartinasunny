# 1. Testing for increase in sales
Hypothesis:
* Null Hypothesis (H0) :- Sales before digital marketing and after digital marketing are the same.

* Alternate Hypothesis (H1 or Ha) :- Sales after digital marketing is greater than the sales before digital marketing .

As the sample size is small(n < 30), we use one tailed t-test to test our hypothesis

```python
import scipy.stats as stats

tscore, pval = stats.ttest_ind(data['Sales_After_digital_add(in $)'],data['Sales_before_digital_add(in $)'],equal_var=True, alternative='greater')
print('pvalue is :', pval)
```

# Conclusion :-
Since the p value is less than 0.05 (Alpha value), we reject the Null Hypothesis. In Short, we can say that there is an increase in Sales after implementing digital marketing.

# 2. The company needs to check whether there is any dependency between the features “Region” and “Manager”.
To check whether there is any dependency between the features “Region” and “Manager” we are setting the below two hypothesis :

* Null Hypothesis (H0) : there is no dependency between the features “Region” and “Manager"
* Alternate Hypothesis (H1 or Ha) : there is significant dependency between the features “Region” and “Manager"

# Insight:-
* Since pValue is higher than alpha (0.05) we conclude that we accept the Null hypothesis.

That means,there is no dependency between the features "Region" and "Manager".
